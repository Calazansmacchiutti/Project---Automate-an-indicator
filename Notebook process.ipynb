# Definir criar uma tabela para cada loja e definir o dia do indicador

```python
#planilha de vendas é diferente da planilha de lojas e uma está inclusa em outra
vendas = vendas.merge(lojas, on='ID Loja')

display(vendas)
```

```python
dicionario_lojas = {}
#cria uma variável para cada nome de loja na lista de lojas [define a coluna]:
for loja in lojas['Loja']:
    dicionario_lojas[loja] = vendas.loc[vendas['Loja']==loja, :]
#testando o código
display(dicionario_lojas['Rio Mar Recife'])
```

```python
indicador_dia = vendas['Data'].max()
print(indicador_dia.year)
print(indicador_dia.day)
print(indicador_dia)
```

# Salvar a planilha na pasta de backup


```python
import pathlib

caminho_backup = pathlib.Path(r"C:/Users/calaz/Documents/[1] - Python DATA/Vscode/1 - Projeto Automação Gerente de Dados/Backup Arquivos Lojas")

arquivos_pasta_backup = caminho_backup.iterdir()
lista_nomes_backup = [arquivo.name for arquivo in arquivos_pasta_backup]
#verificar os arquivos da pasta backup
for loja in dicionario_lojas:
    if loja not in lista_nomes_backup:
       nova_pasta = caminho_backup / loja 
       nova_pasta.mkdir()   
    #definir o nome do arquivo e o local do backup    
    nome_arquivo = '{}_{}_{}.xlsx'.format(indicador_dia.month, indicador_dia.day, loja)
    local_arquivo = caminho_backup / loja / nome_arquivo 
    #usa o pandas para expotar como excel
    dicionario_lojas[loja].to_excel(local_arquivo)
```
